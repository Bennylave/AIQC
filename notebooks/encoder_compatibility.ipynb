```python
for e in encoderz:
    compatibility[str(e)] = []
```

```python
%%capture
for encoder in encoderz:
    try:
        print(str(encoder))
        print(encoder.fit_transform(d1_str).shape)
        print(encoder.fit_transform(d1_str).dtype)
    except:
        print("failed   <----------")
        print("\n\n")
        compatibility[str(encoder)].append(0)
    else:
        print("\n\n")
        compatibility[str(encoder)].append(1)
```
